### `Q-1:` Using `groupby` make groups using the `"Pclass"` column and find out the average age and total number of missing values in the `"Age"` column for every group.

````python
# code here

pclass = titanic_df.groupby('Pclass')
pclass['Age'].mean()
````

````python
titanic_df['Pclass'].unique()
````

````python
l = []
for group,data in pclass:
  df = pd.DataFrame(columns=['Class','total_null_values'])
  df.loc[0,'Class'] = group
  df.loc[0,'total_null_values'] = data['Age'].isnull().sum()
  l.append(df)
pd.concat(l).set_index('Class')
````

### `Q-2:` Using `groupby` make groups using the `"Pclass"` column and fill every group's `"Embarked"` column's missing values with the mode value of that group. After that, print every group's `"Embarked"` column's value counts in ascending order.

````python
# code here
l = []
for group,data in pclass:
  data['Embarked'].fillna(data['Embarked'].value_counts().index[0], inplace=True)


  l.append(data)

titanic_df = pd.concat(l)
````

### `Q-3:` Make groups based on `"Embarked"` column. And for each of this embarked group, make another group based on `"Pclass"` and find out the average fare (round off up to 2 decimal places) for each "Pclass" for each group of "Embarked".

**Sample Output:**

```bash
{'C': {1: 105, 2: 25, 3: 11},
 'Q': {1: 90, 2: 12, 3: 11},
 'S': {1: 70, 2: 20, 3: 15}}
```

````python
# code here

embarked = titanic_df.groupby('Embarked')
````

````python
embarked.groups.keys()
````

````python
l1 = []
for group,data in embarked:
  l1.append(group)
l1
````

````python
l2 = []
for group,data in embarked:
  pclass = data.groupby('Pclass')
  pclass_list = list(pclass.groups.keys())
  mean_fare_list = list(round(pclass['Fare'].mean(),2))
  dic = dict(zip(pclass_list,mean_fare_list))
  l2.append(dic)


l2
````

````python
final = dict(zip(l1,l2))
final
````
